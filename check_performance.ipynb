```python
def information_gain(X_column,y):
    # parent entropy
    parent_entropy = _entropy(y)

    # create children
    left_idxs, right_idxs = (X_column==0).flatten().nonzero()[0], (X_column==1).flatten().nonzero()[0]

    # calculate the weighted avg. entropy of children
    n = len(y)
    n_l, n_r = len(left_idxs), len(right_idxs)
    e_l, e_r = _entropy(y[left_idxs]), _entropy(y[right_idxs])
    
    child_entropy = (n_l/n) * e_l + (n_r/n) * e_r

    # calculate the IG
    print(f'left entropy:{e_l}, right entropy:{e_r}')
    print(f'n_l/n:{n_l/n}, n_r/n:{n_r/n}')
    print(f'parent entropy:{parent_entropy}, child entropy:{child_entropy}')
    information_gain = parent_entropy - child_entropy
    return information_gain

def _entropy(y):
    hist = np.bincount(y)
    ps = hist / len(y)
    return -np.sum([p * np.log(p) for p in ps if p>0])
```
